**Números complejos**

Dado un complejo $z=a+bi$ (en forma binómica),
* Su **módulo** es $$|z|=\sqrt{a^2+b^2}.$$

* Su **argumento** es el ángulo que forma con el semieje positivo horizontal: $$\theta=tan^{-1}(\frac{b}{a}).$$

```python
#from IPython.display import Image
#print("Gráfica de un número complejo z")
#Image(filename='Imagen2.png')
```

**Raíces enésimas de un complejo**

Sean $n$ un número natural y $z$ un complejo, siendo $|z|$ y $\theta$ el módulo y el argumento de $z$, respectivamente. Las raíces n-ésimas de $z$ (o raíces de orden $n$) son $$\sqrt[n]{z}=z^{\frac{1}{n}}=\{|z|^{\frac{1}{n}}e^{\frac{\theta+2\pi k}{n}i}: 0 \leq k \leq n-1 \},$$ o escritas en forma binómica $$\sqrt[n]{z}=|z|^{\frac{1}{n}}(\cos{\frac{\theta+2\pi k}{n}}+i\sin{\frac{\theta+2\pi k}{n}}).$$ El cálculo del conjunto de las raíces n-ésimas de un número complejo $z \in \mathbb{C}$ es conocido como **Teorema de De Moivre**. 

**Ejemplos**

A continuación, calculamos las raíces quintas, cúbicas y cuartas de un número imaginario puro, de un complejo y de un real, respectivamente. 

* Ejemplo 1. Raíces quintas de $z=i$. <br>
El módulo de $z$ es $$|z|=\sqrt{0^2+1^2}=\sqrt{1}=1$$ El argumento de $z$ es $$tan^{-1}(\frac{1}{0})=\frac{\pi}{2}$$ Las raíces quintas de $z$ son $$\sqrt[5]{z}=\{1 \cdot e^{\frac{\frac{\pi}{2}+2\pi k}{5}i}: 0 \leq k \leq 4 \}=\cos{\frac{\frac{\pi}{2}+2\pi k}{5}}+i\sin{\frac{\frac{\pi}{2}+2\pi k}{5}}=Cis(\frac{\frac{\pi}{2}+2\pi k}{5}).$$ Desarrollando la expresión anterior para $k=0,1,2,3,4$ obtendremos los siguientes resultados: $$\sqrt[5]{z}_0=Cis(\frac{\frac{\pi}{2}+2\pi \cdot 0}{5})=Cis(\frac{\pi}{10}); k=0$$ $$\sqrt[5]{z}_1=Cis(\frac{\frac{\pi}{2}+2\pi \cdot 1}{5})=Cis(\frac{\pi}{2})=i; k=1$$ $$\sqrt[5]{z}_2=Cis(\frac{\frac{\pi}{2}+2\pi \cdot 2}{5})=Cis(\frac{9 \pi}{10}); k=2$$ $$\sqrt[5]{z}_3=Cis(\frac{\frac{\pi}{2}+2\pi \cdot 3}{5})=Cis(\frac{13 \pi}{10}); k=3$$ $$\sqrt[5]{z}_4=Cis(\frac{\frac{\pi}{2}+2\pi \cdot 4}{5})=Cis(\frac{17 \pi}{10}); k=4$$ Por lo tanto, el conjunto de raíces quintas de $z=i$ es $$\sqrt[5]{z}=\{Cis(\frac{\pi}{10}), Cis(\frac{\pi}{2}), Cis(\frac{9 \pi}{10}), Cis(\frac{13 \pi}{10}), Cis(\frac{17 \pi}{10}) \}.$$



Ahora escribiremos un programa en Python que encuentre las raíces quintas del número complejo anterior y obtenga su gráfica en el sistema de coordenadas polares.

```python
# Utilizaremos las siguientes bibliotecas de Python:
import math as m
import matplotlib.pyplot as plt
# Dado el número complejo z=0+1i:
z = 0+1j
# El módulo de z es:
moduloz = m.sqrt(0**2+1**2)
# El argumento o ángulo de z es:
#  argz = m.atan(1/0) ¿¿¿???
# Se observa que ya comenzamos con un problema: La función "tangente inversa" de math no acepta la división 
# por cero. Mas aún, ninguna función lo hace.
# Para remediarlo, por ahora, vamos a escribir explícitamente el ángulo como pi/2 radianes:
argz = m.pi/2
# Sólo necesitamos el módulo y ángulo de z para poder graficarlo con ayuda de la siguiente función:
print("GRÁFICA DEL NÚMERO COMPLEJO z=i")
plt.polar([0,argz],[0,moduloz],marker='o')
plt.title("Número complejo $z=i$") # Título de la gráfica.
plt.show() # Mostramos la gráfica de z.
# La primera posición de los corchetes es para establecer el orígen del número complejo respecto a un ángulo y 
# módulo, respectivamente. 
# Es decir, nuestro número complejo tiene como origen las coordenadas polares (0,0) y como extremo final las 
# coordenadas (argz, moduloz).

# PARA LAS RAÍCES:

# El índice n de la raíz es 5:
n = 5      
# Vemos que la fórmula de Moivre está basada en iteraciones con respecto al parámetro k, que varía desde 0 
# hasta n-1.
# Por lo que escribimos un ciclo for para k en un rango de 0 hasta n. Recordemos que si queremos un rango 
# (0,5), Python nos devuelve un rango (0,4). Por tanto,
# para tener un rango (0,n-1) usamos un rango (0,n):

for k in range(0,n):
    # Definimos las coordenadas cartesianas de las raíces de z siguiendo el modelo de Moivre.
    a = moduloz**(1/n)*m.cos((argz+2*m.pi*k)/n)
    b = moduloz**(1/n)*m.sin((argz+2*m.pi*k)/n)
    # Cada raíz es un número complejo:
    raiz = complex(a,b)
    # Mostramos en pantalla cada raíz de z:
    print("La raíz",k,"es:",raiz)
    # Para graficar las raíces, necesitamos sus módulos:
    moduloraiz = m.sqrt(a**2+b**2)
    # y sus argumentos o ángulos:
    # argraiz = ¿¿¿¿¿????? 
    # SURGE UN GRAN PROBLEMA, la función m.atan no devuelve ángulos fuera del intervalo (-90°, 90°). 
                                   # VER SECCIÓN "Función tangente inversa".
    
    # Ya definida la función "tangente inversa", la aplicamos para obtener los ángulos de las raíces:
    argraiz = anguloz(a,b)
    
    # Graficamos cada una de las raíces de z:
    plt.polar([0,argraiz],[0,moduloraiz],marker='o',color='navy') # La sintaxis es muy intuitiva, como ya 
                                                                  # se había explicado antes.
    plt.title("Raíces n-ésimas de $z=i$") # Título de la gráfica.
```

Nótese que al calcular 5 raíces quintas hemos obtenido los vértices de un pentágono centrado en el origen del plano polar. 

**Función tangente inversa** <br>
Como podemos observar, acaba de surgir un problema importante que impide graficar las 5 raíces de $z$. Dado que la función "atan" (tangente inversa) de la biblioteca *math* no devuelve valores de $\theta$ fuera del intervalo $[-\frac{\pi}{2}, \frac{\pi}{2}]$, nos vemos en la necesidad de construir uns función que sí lo haga. Para ello, recordemos los signos de la tangente en cada cuadrante: 
* Cuadrante I: $\sin{\theta}>0 \land \cos{\theta}>0 \implies \tan{\theta}>0$.
* Cuadrante II: $\sin{\theta}>0 \land \cos{\theta}<0 \implies \tan{\theta}<0$.
* Cuadrante III: $\sin{\theta}<0 \land \cos{\theta}<0 \implies \tan{\theta}>0$.
* Cuadrante IV: $\sin{\theta}<0 \land \cos{\theta}>0 \implies \tan{\theta}<0$. <br>

Además, sabemos que $\sin{\theta}$ y $\cos{\theta}$ están directamente relacionados con la parte imaginaria y real de $z$, respectivamente; pues $Im(z)=|z|\sin{\theta}$ y $Re(z)=|z|\cos{\theta}$. Una vez dicho lo anterior, ya tenemos todo lo necesario para definir la función que llamaremos *anguloz(x,y)*.

```python
def anguloz(x,y):
    # Definimos los valores del ángulo para un complejo sobre cada uno de los 4 semiejes del plano cartesiano:
    if x>0 and y==0:          # Semieje x positivo.
        angz = 0
    if x==0 and y>0:          # Semieje y positivo.
        angz = m.pi/2
    if x<0 and y==0:          # Semieje x negativo.
        angz = m.pi
    if x==0 and y<0:          # Semieje y negativo.
        angz = (3*m.pi)/2
   
    # Ahora definimos el ángulo para cada uno de los cuadrantes:

     # Cuadrante I: tangente>0, no hay problema (la función "atan" sirve).
    if x>0 and y>0:                       
        angz = m.atan(y/x)
    # Cuadrante II: tangente<0, habrá que sumar 180° al ángulo devuelto por "atan".    
    if x<0 and y>0:                        
        angz = m.atan(y/x) + m.pi
    # Cuadrante III: tangente>0, habrá que sumar 180° al ángulo devuelto por "atan".    
    if x<0 and y<0:                       
        angz = m.atan(y/x) + m.pi
    # Cuadrante IV: tangente<0, habrá que sumar 360° al ángulo devuelto por "atan".    
    if x>0 and y<0:                       
        angz = m.atan(y/x) + 2*m.pi
    return angz    # Regresa el valor del ángulo ajustado a su cuadrante correspondiente según "x" y "y".
```

Una vez definida la función, ya podemos regresar al código de antes y aplicarla a los valores de $a$ y $b$ correspondientes a la parte real e imaginaria de las raíces de $z$, respectivamente. 

* Ejemplo 2. Raíces cúbicas de $z=1+i$. <br>
El módulo de $z$ es $$|z|=\sqrt{1^2+1^2}=\sqrt{2}$$ El argumento de $z$ es $$tan^{-1}(\frac{1}{1})=\frac{\pi}{4}$$ Las raíces cúbicas de $z$ son $$\sqrt[3]{z}=\{\sqrt[6]{2} \cdot e^{\frac{\frac{\pi}{4}+2\pi k}{3}i}: 0 \leq k \leq 2 \}=Cis(\frac{\frac{\pi}{4}+2\pi k}{3}).$$ Desarrollando la expresión anterior para $k=0,1,2$ obtendremos los siguientes resultados: $$\sqrt[3]{z}_0=\sqrt[6]{2}Cis(\frac{\frac{\pi}{4}+2\pi \cdot 0}{3})=\sqrt[6]{2}Cis(\frac{\pi}{12}); k=0$$ $$\sqrt[3]{z}_1=\sqrt[6]{2}Cis(\frac{\frac{\pi}{4}+2\pi \cdot 1}{3})=\sqrt[6]{2}Cis(\frac{3 \pi}{4}); k=1$$ $$\sqrt[3]{z}_2=\sqrt[6]{2}Cis(\frac{\frac{\pi}{4}+2\pi \cdot 2}{3})=\sqrt[6]{2}Cis(\frac{17 \pi}{12}); k=2$$ Por lo tanto, el conjunto de raíces cúbicas de $z=1+i$ es $$\sqrt[3]{z}=\{\sqrt[6]{2}Cis(\frac{\pi}{12}), \sqrt[6]{2}Cis(\frac{3 \pi}{4}), \sqrt[6]{2}Cis(\frac{17 \pi}{12})\}.$$


Para agilizar el proceso de encontrar las raíces n-ésimas de un número complejo $z=a+bi$, podemos convertir el código anterior en una función que dependa de 3 parámetros: la parte real e imaginaria de $z$ y el índice de radicación $n \in \mathbb{N}$ mayor o igual que 2.

```python
def raicesz(a,b,n):                                                    # La función depende de a, b y n.
    moduloz = m.sqrt(a**2+b**2)                                        # Módulo del complejo z.
    print("El módulo de z, es:",moduloz,"unidades.")                   # Se muestra el valor del módulo.
    argz = anguloz(a,b)                          # Ángulo del complejo definido con la función previa "anguloz".
    print("El argumento o ángulo de z es:",argz,"radianes.")           # Se muestra el argumento de z.
    z = complex(a,b)                                   
    print("El número complejo z es:",z)                     
    print("Gráfica del número complejo z")
    plt.polar([0,argz],[0,moduloz],marker='o',color='navy')            # Gráfica de z en el plano polar.
    plt.title("Número complejo $z=a+bi$")                              # Título de la gráfica.
    plt.show()                                                         # Se muestra primero esta gráfica.
    for k in range(0,n):                                               # Realmente es el rango (0,n-1).
        c = moduloz**(1/n)*m.cos((argz+2*m.pi*k)/n)                    # Parte real de las raíces.
        d = moduloz**(1/n)*m.sin((argz+2*m.pi*k)/n)                    # Parte imaginaria de las raíces.
        raiz = complex(c,d)
        print("La raíz",k,"es:",raiz)
        moduloraiz = m.sqrt(c**2+d**2)                                 # Módulo de las raíces.
        argraiz = anguloz(c,d)                                         # Argumento de las raíces.
        plt.polar([0,argraiz],[0,moduloraiz],marker='o',color='navy')  # Gráfica de cada una de las n raíces.
        plt.title("Raíces n-ésimas de $z$") 
```

Podemos resolver el ejemplo 2 aplicando directamente la función definida "raicesz(a,b,n)" de la siguiente forma: <br>
Sea $z \in \mathbb{C}$ dado por $z=1+i$. La parte real es $Re(z)=a=1$, la parte imaginaria es $Im(z)=b=1$ y el índice de la raíz es $n=3$. Por lo tanto:

```python
raicesz(1,1,3)
```

Ahora obtuvimos los tres vértices de un trángulo, al calcular 3 raíces cúbicas.

* Ejemplo 3. Raíces cuartas de $z=1$. <br>
Basta con identificar $Re(z)=1$, $Im(z)=0$ y el índice de la raíz $n=4$. De esta manera, tendremos que:

```python
raicesz(1,0,4)
```

Seguimos con el mismo patrón: al calcular 4 raíces cuartas, obtenemos los 4 vértices de un cuadrado centrado en el origen del plano polar.

**Raíces enésimas de cualquier número complejo ingresado**

Ya que revisamos ejemplos particulares de cómo calcular las raíces enésimas de un número complejo, ahora apuntaremos a contruir un programa más interactivo con el usuario; es decir, se hallarán y graficarán las raíces de cualquier número $z \in \mathbb{C}$ definido por $z=a+bi$ introducido por el usuario. Además, tendremos la oportunidad de guardar en una lista los valores tanto de la parte real como imaginaria de todas las raíces del complejo introducido. Esto con intención de visualizar otras gráficas que también son interesantes de analizar como por ejemplo *Parte imaginaria VS Parte real*, *Parte real VS argraiz* y *Parte imaginaria VS argraiz*. 

**Función "raices de z binómica"**

```python
                             # FUNCIÓN "raícesz" en términos de a, b y n. FORMA BINÓMICA.
# Esta función es muy similar a la anterior, lo único que hemos cambiado es que ahora agregaremos la indicación 
# de agregar a 3 listas los elementos que se obtengan de c, d y argz.
def raiceszbinomica(a,b):                                                      
    moduloz = m.sqrt(a**2+b**2)                                          
    print("El módulo de z, es:",moduloz,"unidades.")                     
    argz = anguloz(a,b)                                                   
    print("El argumento o ángulo de z es:",argz,"radianes.")             
    z = complex(a,b)                                   
    print("El número complejo z es:",z)
    print()
    print("Gráfica del número complejo z")
    plt.polar([0,argz],[0,moduloz],marker='o',color='navy')               
    plt.title("Número complejo $z=a+bi$")                                
    plt.show() 

    n = int(input("Ingrese el índice de la raíz n:" )) # El usuario ingresa el índice de la raíz.
    
    for k in range(0,n):        
        c = moduloz**(1/n)*m.cos((argz+2*m.pi*k)/n)                       
        d = moduloz**(1/n)*m.sin((argz+2*m.pi*k)/n)
        # Agregamos todos los elementos "c" a la lista "parte real" de las raíces.
        partereal.append(c)
        # Agregamos todos los elementos "d" a la lista "parte imaginaria" de las raíces.
        parteimaginaria.append(d) 
        raiz = complex(c,d)
        print("La raíz",k,"es:",raiz)
        moduloraiz = m.sqrt(c**2+d**2)                                    
        argraiz = anguloz(c,d)    
        argumentoraiz.append(argraiz)      # Agregamos todos los elementos "argraiz" a la lista "argumentoraiz".
        plt.polar([0,argraiz],[0,moduloraiz],marker='o',color='navy')     
        plt.title("Raíces n-ésimas de $z$") 
```

**Número ingresado en forma binómica**

```python
                                                     # FORMA BINÓMICA
print("Bienvenido. Este programa dibuja un círculo de radio |z|^{1/n} utilizando como entrada un número complejo z. Considere que n deberá tender a infinito.")
print("escrito ya sea en su forma binómica o en su forma polar.")
print()
print("Además de ingresar el número complejo, introducirá el índice de la raíz que quiere encontrar para generar el círculo.")
print("También es posible obtener tres gráficas diferentes con sus respectivos archivos de datos.")
print()
partereal = []          # Lista vacía en donde se guardarán los valores de las partes reales de las raíces.
parteimaginaria = []    # Lista vacía en donde se guardarán los valores de las partes imaginarias de las raíces.
argumentoraiz = []      # Lista vacía en donde se guardarán los valores de los ángulos de las raíces.

a = float(input("Ingrese el valor de la parte real de z, a:" ))      #El usuario ingresa valores flotantes.
b = float(input("Ingrese el valor de la parte imaginaria de z, b:" ))
print()
raiceszbinomica(a,b) # Invocando la ejecución de la función raiceszbinomica(a,b).
```

Al calcular las primeras 1000 raíces del número complejo $z=1$ obtenemos una figura geométrica muy parecida a un círculo de radio $r=1$, el cual coincide con el módulo de $z$. Sin embargo, el módulo de $z$ no siempre es igual al radio del círculo que se genera, sino que $r=\sqrt[1000]{|z|}$. **Número ingresado en forma polar**

```python
                                                      # FORMA POLAR
partereal = []               
parteimaginaria = []        
argumentoraiz = []     
# En lugar de ingresar la parte real, el usuario ingresa el módulo del complejo.
r = float(input("Ingrese el valor del módulo de z, r: "))
theta = float(input("Ingrese el argumento de z en radianes: "))
# Para usar la misma función que ya definimos, habrá que calcular la parte real e imaginaria de z, 
# que como ya mencionamos, equivalen a:
a = r*m.cos(theta)                         
b = r*m.sin(theta)
# Aplicamos la función que hemos definido a los valores a y b provenientes del módulo y ángulo de z.
raiceszbinomica(a,b)
```

Con el número complejo $z=(5, \frac{1}{2} rad)$ (en coordenadas polares) hemos generado algo parecido a un círculo de radio $r=\sqrt[1000]{5}$. Esto nos lleva a pensar definitivamente que un círculo se puede formar con las raíces enésimas de cualquier $z \in \mathbb{C}$, y tendrá un radio del mismo tamaño que el módulo de dicho número complejo.

**Tres gráficas ilustrativas**

* Gráfica 1. Parte imaginaria VS Parte real <br>
Recordemos que la parte real de $z$ está dada por $Re(z)=|z|\cos(\theta)$ y la parte imaginaria por $Im(z)=|z|\sin(\theta)$. Así, cuando graficamos $Im(raiz) \ vs \ Re(raiz)$ estamos construyendo la relación $Im(raiz)[(Re(raiz)]$, la cual se puede comparar con las coordenadas binómicas de todas las raíces que hemos calculado; esto es, aquellos pares ordenados $(Re(raiz), Im(raiz))$. Por lo tanto, obtendremos un lugar geométrico que unirá todos los puntos (coordenadas binómicas) que hayamos generado al calcular las raíces de orden $n$ del número complejo $z$.

* Gráfica 2. Parte real VS argraiz <br>
Si entendimos bien la primera gráfica, la segunda y la tercera no causarán ningún problema. Graficar la parte real de las raíces en relación al argumento de las mismas, significa que obtendremos un lugar geométrico de comportamiento cosenoidal, pues recordemos que la parte real de $z$ está modulada por $|z|\cos(\theta)$, es decir, tendremos una **función** cosenoidal que precisamente depende del ángulo o argumento $\theta$. 

* Gráfica 3. Parte imaginaria VS argraiz <br>
Entonces, esta gráfica tendrá un comportamiento senoidal debido a que la parte imaginaria de un complejo $z$ como función de su argumento $\theta$ está modulada por  $|z|\sin(\theta)$.

Llegado hasta este punto, ya tenemos todas las herramientas tanto algebraicas como gráficas para enunciar la siguiente "definición" de círculo vista desde el estudio de los números complejos: <br>

**Círculo: Lugar geométrico del conjunto de todas las raíces enésimas de un número complejo $z$ de módulo $|z|$ y argumento $\theta$.**

Esto es, dado el número complejo $z=(|z|, \theta)$ (en coordenadas polares), el lugar geométrico conformado por el conjunto de todas sus raíces enésimas ($n \geq 2$) $$\sqrt[n]{z}=z^{\frac{1}{n}}=\{|z|^{\frac{1}{n}}e^{\frac{\theta+2\pi k}{n}i}: 0 \leq k \leq n-1 \}$$ es un círculo de radio $r=\sqrt[n]{|z|}$.
Como nota importante, debemos observar que a medida que $n$ aumenta, el módulo de las raíces irá tendiendo a 1. Esto quiere decir que estamos generando un círculo de radio 1 cuando $n \to \infty$, pues $$\lim_{n \to \infty} \sqrt[n]{a}=1,\forall a \in \mathbb{R}$$ De esta manera, tenemos una visión diferente del círculo unitario como $$C_1=\lim_{n \to \infty} \{\sqrt[n]{z}\}$$ donde $z$ es cualquier número complejo.
