```python
import math as m
import matplotlib.pyplot as plt

def anguloz(x,y):
    # Definimos los valores del ángulo para un complejo sobre cada uno de los 4 semiejes del plano cartesiano:
    if x>0 and y==0:          # Semieje x positivo.
        angz = 0
    if x==0 and y>0:          # Semieje y positivo.
        angz = m.pi/2
    if x<0 and y==0:          # Semieje x negativo.
        angz = m.pi
    if x==0 and y<0:          # Semieje y negativo.
        angz = (3*m.pi)/2
   
    # Ahora definimos el ángulo para cada uno de los cuadrantes:

     # Cuadrante I: tangente>0, no hay problema (la función "atan" sirve).
    if x>0 and y>0:                       
        angz = m.atan(y/x)
    # Cuadrante II: tangente<0, habrá que sumar 180° al ángulo devuelto por "atan".    
    if x<0 and y>0:                        
        angz = m.atan(y/x) + m.pi
    # Cuadrante III: tangente>0, habrá que sumar 180° al ángulo devuelto por "atan".    
    if x<0 and y<0:                       
        angz = m.atan(y/x) + m.pi
    # Cuadrante IV: tangente<0, habrá que sumar 360° al ángulo devuelto por "atan".    
    if x>0 and y<0:                       
        angz = m.atan(y/x) + 2*m.pi
    return angz    # Regresa el valor del ángulo ajustado a su cuadrante correspondiente según "x" y "y".

                             # FUNCIÓN "raícesz" en términos de a, b y n. FORMA BINÓMICA.
# Esta función es muy similar a la anterior, lo único que hemos cambiado es que ahora agregaremos la indicación 
# de agregar a 3 listas los elementos que se obtengan de c, d y argz.
def raiceszbinomica(a,b):                                                      
    moduloz = m.sqrt(a**2+b**2)                                          
    print("El módulo de z, es:",moduloz,"unidades.")                     
    argz = anguloz(a,b)                                                   
    print("El argumento o ángulo de z es:",argz,"radianes.")             
    z = complex(a,b)                                   
    print("El número complejo z es:",z)
    print()
    print("Gráfica del número complejo z")
    plt.polar([0,argz],[0,moduloz],marker='o',color='navy')               
    plt.title("Número complejo $z=a+bi$")                                
    plt.show() 

    n = int(input("Ingrese el índice de la raíz n:" )) # El usuario ingresa el índice de la raíz.
    
    for k in range(0,n):        
        c = moduloz**(1/n)*m.cos((argz+2*m.pi*k)/n)                       
        d = moduloz**(1/n)*m.sin((argz+2*m.pi*k)/n)
        # Agregamos todos los elementos "c" a la lista "parte real" de las raíces.
        partereal.append(c)
        # Agregamos todos los elementos "d" a la lista "parte imaginaria" de las raíces.
        parteimaginaria.append(d) 
        raiz = complex(c,d)
        print("La raíz",k,"es:",raiz)
        moduloraiz = m.sqrt(c**2+d**2)                                    
        argraiz = anguloz(c,d)    
        argumentoraiz.append(argraiz)      # Agregamos todos los elementos "argraiz" a la lista "argumentoraiz".
        plt.polar([0,argraiz],[0,moduloraiz],marker='o',color='navy')     
        plt.title("Raíces n-ésimas de $z$") 


print("Bienvenido. Este programa dibuja un círculo de radio r utilizando como entrada un número complejo z escrito ya sea en su forma binómica o en su forma polar")
print("Además de ingresar el número complejo, introducirá el índice de la raíz que quiere encontrar para generar el círculo.")
print("También es posible obtener tres gráficas diferentes con sus respectivos archivos de datos.")
print()
#partereal = []               # Lista vacía en donde se guardarán los valores de las partes reales de las raíces.
#parteimaginaria = []         # Lista vacía en donde se guardarán los valores de las partes imaginarias de las raíces.
#argumentoraiz = []           # Lista vacía en donde se guardarán los valores de los ángulos de las raíces.

respuesta = True
while respuesta==True:
    respuesta = input("¿En qué forma prefiere ingresar el número complejo z? binómica/polar: ")
    
    if respuesta == 'binómica':
        respuesta = False
        partereal = []               # Lista vacía en donde se guardarán los valores de las partes reales de las raíces.
        parteimaginaria = []         # Lista vacía en donde se guardarán los valores de las partes imaginarias de las raíces.
        argumentoraiz = []           # Lista vacía en donde se guardarán los valores de los ángulos de las raíces.
        print()
        a = float(input("Ingrese el valor de la parte real de z, a:" ))
        b = float(input("Ingrese el valor de la parte imaginaria de z, b:" ))
        print()
        raiceszbinomica(a,b)
        
        respuesta  = input("¿Desea generar archivos.txt para almacenar los valores obtenidos? sí/no:" )
        if respuesta == 'no':
            respuesta = False
            print()
            print("Si lo desea, puede obtener alguna de las siguientes gráficas:")
            print("1. Parte imaginaria VS Parte real.")
            print("2. Parte real VS argraiz.")
            print("3. Parte imaginaria VS argraiz.")
            print("4. Salir del programa.")
            respuesta = input("Seleccione alguna opción: ")
            if respuesta=='1':
                respuesta = True
                plt.plot(partereal,parteimaginaria)
                plt.title("$Im(raiz)$ vs $Re(raiz)$") 
                plt.xlabel("$Re(raiz)$")                      
                plt.ylabel("$Im(raiz)$")
                plt.show()
            elif respuesta=='2':
                respuesta = True
                plt.plot(argumentoraiz,partereal)
                plt.title("$Re(raiz)$ vs $arg(raiz)$") 
                plt.xlabel("$arg(raiz)$")                      
                plt.ylabel("$Re(raiz)$")
                plt.show()
            elif respuesta=='3':
                respuesta = True
                plt.plot(argumentoraiz,parteimaginaria)
                plt.title("$Im(raiz)$ vs $arg(raiz)$") 
                plt.xlabel("$arg(raiz)$")                      
                plt.ylabel("$Im(raiz)$")
                plt.show()
            else:
                print("Gracias por reproducir este programa.")
                break
        if respuesta=='sí':
            respuesta = False
            Real = open("ParteReal.txt","w")
            Real.writelines("%s\n" % s for s in list(partereal))
            Real.close()
            print("Se ha generado el archivo ParteReal.txt correctamente.")
            print()
            Imaginario = open("ParteImaginaria.txt","w")
            Imaginario.writelines("%i\n" % i for i in list(parteimaginaria))
            Imaginario.close
            print("Se ha generado el archivo ParteImaginaria.txt correctamente.")
            print()
            Anguloraiz = open("ArgumentoRaiz.txt","w")
            Anguloraiz.writelines("%u\n" % u for u in list(argumentoraiz))
            Anguloraiz.close
            print("Se ha generado el archivo ArgumentoRaiz.txt correctamente.")
            
            print()
            print("Si lo desea, ahora puede obtener alguna de las siguientes gráficas:")
            print("1. Parte imaginaria VS Parte real.")
            print("2. Parte real VS argraiz.")
            print("3. Parte imaginaria VS argraiz.")
            print("4. Salir del programa.")
            print()
            respuesta = input("Seleccione alguna opción: ")
            if respuesta=='1':
                respuesta = True
                plt.plot(partereal,parteimaginaria)
                plt.title("$Im(raiz)$ vs $Re(raiz)$") 
                plt.xlabel("$Re(raiz)$")                      
                plt.ylabel("$Im(raiz)$")
                plt.show()
            elif respuesta=='2':
                respuesta = True
                plt.plot(argumentoraiz,partereal)
                plt.title("$Re(raiz)$ vs $arg(raiz)$") 
                plt.xlabel("$arg(raiz)$")                      
                plt.ylabel("$Re(raiz)$")
                plt.show()
            elif respuesta=='3':
                respuesta = True
                plt.plot(argumentoraiz,parteimaginaria)
                plt.title("$Im(raiz)$ vs $arg(raiz)$") 
                plt.xlabel("$arg(raiz)$")                      
                plt.ylabel("$Im(raiz)$")
                plt.show()
                print("Gracias por reproducir este programa.")
                break
            
                
        
    elif respuesta == 'polar':
        respuesta = False
        partereal = []               # Lista vacía en donde se guardarán los valores de las partes reales de las raíces.
        parteimaginaria = []         # Lista vacía en donde se guardarán los valores de las partes imaginarias de las raíces.
        argumentoraiz = []           # Lista vacía en donde se guardarán los valores de los ángulos de las raíces.
        r = float(input("Ingrese el valor del módulo de z, r: "))
        theta = float(input("Ingrese el argumento de z en radianes: "))
        a = r*m.cos(theta)
        b = r*m.sin(theta)
        print()
        raiceszbinomica(a,b)
        
        respuesta  = input("¿Desea generar archivos.txt para almacenar los valores obtenidos? sí/no:" )
        if respuesta == 'no':
            respuesta = False
            print()
            print("Si lo desea, puede obtener alguna de las siguientes gráficas:")
            print("1. Parte imaginaria VS Parte real.")
            print("2. Parte real VS argraiz.")
            print("3. Parte imaginaria VS argraiz.")
            print("4. Salir del programa.")
            respuesta = input("Seleccione alguna opción: ")
            if respuesta=='1':
                respuesta = True
                plt.plot(partereal,parteimaginaria)
                plt.title("$Im(raiz)$ vs $Re(raiz)$") 
                plt.xlabel("$Re(raiz)$")                      
                plt.ylabel("$Im(raiz)$")
                plt.show()
            elif respuesta=='2':
                respuesta = True
                plt.plot(argumentoraiz,partereal)
                plt.title("$Re(raiz)$ vs $arg(raiz)$") 
                plt.xlabel("$arg(raiz)$")                      
                plt.ylabel("$Re(raiz)$")
                plt.show()
            elif respuesta=='3':
                respuesta = True
                plt.plot(argumentoraiz,parteimaginaria)
                plt.title("$Im(raiz)$ vs $arg(raiz)$") 
                plt.xlabel("$arg(raiz)$")                      
                plt.ylabel("$Im(raiz)$")
                plt.show()
            else:
                print("Gracias por reproducir este programa.")
                break
        if respuesta=='sí':
            respuesta = False
            Real = open("ParteReal.txt","w")
            Real.writelines("%s\n" % s for s in list(partereal))
            Real.close()
            print("Se ha generado el archivo ParteReal.txt correctamente.")
            print()
            Imaginario = open("ParteImaginaria.txt","w")
            Imaginario.writelines("%i\n" % i for i in list(parteimaginaria))
            Imaginario.close
            print("Se ha generado el archivo ParteImaginaria.txt correctamente.")
            print()
            Anguloraiz = open("ArgumentoRaiz.txt","w")
            Anguloraiz.writelines("%u\n" % u for u in list(argumentoraiz))
            Anguloraiz.close
            print("Se ha generado el archivo ArgumentoRaiz.txt correctamente.")
            print()
            print("Si lo desea, ahora puede obtener alguna de las siguientes gráficas:")
            print("1. Parte imaginaria VS Parte real.")
            print("2. Parte real VS argraiz.")
            print("3. Parte imaginaria VS argraiz.")
            print("4. Salir del programa.")
            print()
            respuesta = input("Seleccione alguna opción: ")
            if respuesta=='1':
                respuesta = True
                plt.plot(partereal,parteimaginaria)
                plt.title("$Im(raiz)$ vs $Re(raiz)$") 
                plt.xlabel("$Re(raiz)$")                      
                plt.ylabel("$Im(raiz)$")
                plt.show()
            elif respuesta=='2':
                respuesta = True
                plt.plot(argumentoraiz,partereal)
                plt.title("$Re(raiz)$ vs $arg(raiz)$") 
                plt.xlabel("$arg(raiz)$")                      
                plt.ylabel("$Re(raiz)$")
                plt.show()
            elif respuesta=='3':
                respuesta = True
                plt.plot(argumentoraiz,parteimaginaria)
                plt.title("$Im(raiz)$ vs $arg(raiz)$") 
                plt.xlabel("$arg(raiz)$")                      
                plt.ylabel("$Im(raiz)$")
                plt.show()
            else:
                print("Gracias por reproducir este programa.")
                break
            
            
    else: 
        print("Ingrese una opción válida.")
    
```
