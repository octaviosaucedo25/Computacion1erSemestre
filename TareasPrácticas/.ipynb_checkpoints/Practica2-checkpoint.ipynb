*Normalmente*, si sólo ejecutamos el comando _cat_ con el archivo names_2009-08-30.txt, nos enlista su contenido en la salida predeterminada, es decir, en la pantalla de la terminal:

```python
%%bash
cd Resultado
cat names_2009-08-30.txt
```

Al ejecutar el comando del inciso _a_, el comando _cat_ redirecciona su salida al archivo names_cat.txt, el cual no existía, así que lo crea; y al mismo tiempo está realizando una copia del contenido del archivo names_2009-08-30-txt al archivo que sea ha creado.

*Nota*: Si el segundo archivo ya existía, el nuevo contenido se sobreescribe en él y su contenido anterior se pierde.  

```python
Image(filename='Imagen2.png')
```

Mientras que con el comando del inciso _b_, sucede lo siguiente: como el archivo names_cat.txt ya existía, entonces el nuevo contenido que le manda _cat_ se añade al final de su contenido original sin que se pierda. Por lo tanto, debido al comando ejecutado anteriormente, tendremos dos veces el mismo contenido del archivo names_2009-08-30.txt en names_cat.txt

```python
%%bash
cd Resultado
cat names_cat.txt
```

*Nota*: el comando >> se trata de un redireccionamiento no destructivo o de adición. 

4. ¿Cuál es la diferencia entre estos comandos?
 
    a) `cat names.txt 2> err.dat`
    
    b) `cata names.txt 2> err.dat`
    

```python
Image(filename='Imagen4.png')
```

En este caso, la salida del comando _cat_ se redirige a un archivo DAT (o ficheros de datos); los cuales contienen datos genéricos que pueden ser utilizados o indexados por otros programas. Entonces surgen las siguientes dos situaciones:
1. El redireccionamiento _2>_ manda la salida del comando _cat_ al archivo _err.dat_ donde el _error estándar_ es que no existe el archivo o directorio _names.txt_ 
2. Se trata de la misma situación anterior, pero ahora el _error estándar_ es que no se encontró el comando _cata_. 

5. Por último ejecuta: `cat > feliz` enseguida escibre algo y luego ejecuta `Ctrl+D` dos veces. ¿qué fue lo que sucedio?

```python
Image(filename='Imagen5.png')
```

*OBSERVACIONES:*
1. Al ejecutar _cat > feliz_ suceden dos cosas: se crea el archivo _feliz_ (no existía) y puedo escribir su contenido en la misma pantalla de la terminal. Así, al abrir el archivo _feliz_ se encuentra lo que escribí anteriormente.
2. Al ejecutar otra vez _cat > feliz_ y escribir otro contenido, este se sobreescribe en el anterior y lo elimina. 
3. Al ejecutar _cat >> feliz_ vuelvo a escribir otro texto, pero esta vez se añade después del contenido anterior en lugar de eliminarlo. Esto es debido a que utilicé el redireccionamiento aditivo.

## Permisos

12. Explica que hace el comando `umask` y pon un ejemplo de su uso. 

El comando umask, es la abreviatura de user file-creation mode mask, y sirve para establecer los permisos por defecto que tendrán los nuevos ficheros y directorios que creemos. Utiliza algo llamado _máscaras_ escritas por método octal.

* La máscara para un directorio es _777_, y si queremos establecer los persmisos _755_ a un directorio que crearemos, hay que restar los valores de los persmisos a los de la máscara, es decir, escribimos _umask 022_. Luego de eso, creamos el directorio nuevo con _mkdir_.

* La máscara para un archivo es _666_, y si queremos establecer los persmisos _644_ a un archivo que crearemos, hay que restar los valores de los persmisos respectivamente con los de la máscara, es decir, escribimos _umask 022_. Luego de eso, creamos el archivo nuevo con _touch_.

Ejemplo: Dentro del directorio _Resultado_ crearemos un archivo llamado _permisosumask.txt_, le asignaremos permisos de lectura y ejecución a todos los usuarios, es decir, los permisos _555_. Para ello, escribimos primero _umask 222_ y enseguida creamos el nuevo archivo.

```python
%%bash
pwd
cd Resultado
pwd
ls -l
umask
umask 222
touch permisosumask.txt
ls -l permisosumask.txt
```
